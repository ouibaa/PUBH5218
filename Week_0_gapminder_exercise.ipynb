```
# import libraries
install.packages("skimr")
library(skimr)
```

```
# Read data
data = read.csv("./gapminder_data.csv")
skim(data)
```

# Goal figure to produce
x-axis is `health_spend_per_person`

y-axis is `life_expectancy`

size of the circle is `population`

colour of circle is `continent`

```
# Load ggplot2
library(ggplot2)
```

```
ggplot(data=data, aes(x=health_spend_per_person, y=life_expectancy)) +
  geom_point(aes(size = population, colour=continent)) +
  theme_minimal() +
  labs(title = "Life expectancy vs. Total health spending per person", x = "Total health spending per person", y = "Life expectancy (years)")
```
