# Week 0: Creating graphs with ggplot

gapminder.csv dataset visualisation in ggplot2

ggplot = "grammar of graphics" plot

Tutorial example

```
#load libraries
install.packages("skimr")
install.packages("tidyverse")
library(skimr)
library(tidyverse)
```

```
# Import dataset as data
data = read.csv("./bmd_restricted.csv")
skim(data)
```

```
# Weight histogram
ggplot( data = data, aes(x = weight_kg))
# aes = aesthetic
# histogram - x is spread/values, y is frequency
# Generates an empty plot, as we haven't told the graph how our data should appear
# It's determined the upper and lower limits of the weight variable
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram()
# takes the weight_kg values and makes a geometric histogram
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram( binwidth = 3)
  # makes the width of each binto be 3 units wide
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram( binwidth = 3, colour = "black", fill = "lightblue")
  # Changes the colours and fills
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram( binwidth = 3, colour = "black", fill = "lightblue") +
  labs(x = "Weight (kg)", title = "Histogram of Age") #Added labels
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram( binwidth = 3, colour = "black", fill = "lightblue") +
  labs(x = "Weight (kg)", title = "Histogram of Age") +
  theme_minimal() #Changes the background to minimal
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram( binwidth = 3, colour = "black", fill = "lightblue") +
  labs(x = "Weight (kg)", title = "Histogram of Age") +
  theme_minimal() + 
  facet_wrap(~sex) # If data was categorical and you want to separate the graph based on categories
```

```
data$sex <- factor(data$sex,
  levels = c("F", "M"),
  labels = c("Female", "Male"))
# Convert the character variables to factor variables
```

```
ggplot(data = data, aes(x = weight_kg)) + 
  geom_histogram( binwidth = 3, colour = "black", fill = "lightblue") +
  labs(x = "Weight (kg)", title = "Histogram of Age") +
  theme_minimal() + 
  facet_wrap(~sex)
```

```
ggplot(data = data, aes(x = weight_kg, y = height_cm)) #empty graph (no geoms provided yet)
```

```
ggplot(data = data, aes(x = weight_kg, y = height_cm)) + 
  geom_point(aes(colour = sex))+ #Aes means we are mapping some variable to values in our plot
  theme_minimal()
```

```
ggplot(data = data, aes(x = weight_kg, y = height_cm)) + 
  geom_point(aes(colour = sex))+ #Aes means we are mapping some variable to values in our plot
  theme_minimal()+
  scale_colour_manual(values = c("darkgreen", "lightblue")) #Override default colours from ggplot
  # R can also recognise hex codes
```
